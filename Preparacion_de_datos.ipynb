# **5. Limpieza de datos nulos: Imputación**

Estrategia:
* Eliminar columnas con mas de 15%-20% de nulos
* Imputar por media, moda, mediana, vecinos cercanos. No se puede imputar más allá del 15% de los datos.



* Carbón_t has 5 (3.6%) missing values
* Energía Nuclear_t has 34 (24.8%) missing values
* Energía Hidroeléctrica_t has 13 (9.5%) missing values

```
#Energía Nuclear_t tiene más del 15% de nulos, por lo tanto se elimina
data = data.drop(columns='Energía Nuclear_t')
data.head()
```

```
8#Limpieza de datos nulos: Imputación por la media y moda
from sklearn.impute import SimpleImputer

#Imputacion de variables numéricas: media, mediana
imputar_media = ['Gas Natural_t','Carbón_t','Energía Hidroeléctrica_t','Energías Renovables_t']
ImpNumeros = SimpleImputer(missing_values=np.nan, strategy='mean')
data[imputar_media] = ImpNumeros.fit_transform(data[imputar_media])

#Imputacion de variables categóricas: moda
imputar_moda = ['Region']
ImpCategorias = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
data[imputar_moda] = ImpCategorias.fit_transform(data[imputar_moda])

data.info()
```

```
#Valores de la imputación
print(ImpNumeros.statistics_) #['Gas Natural_t','Carbón_t','Energía Hidroeléctrica_t','Energías Renovables_t']
print(ImpCategorias.statistics_) #['Region']
```

```
#Guardamos los datos limpios
data.to_excel('./datos_limpios.xlsx')
```

# 9. Transformaciones

Se sugiere realizar estas operaciones en el notebook donde se realice la creación de los modelos:
- Normalizar las variables numéricas
- Crear dummies





```
data.head()
```

```
#Normalización: Sólo para variables numéricas (no dummies)
from sklearn.preprocessing import MinMaxScaler

variables_predictoras_numericas=['Petróleo_t','Gas Natural_t','Carbón_t','Energía Hidroeléctrica_t','Energías Renovables_t']

min_max_scaler = MinMaxScaler()
data[variables_predictoras_numericas] = min_max_scaler.fit_transform(data[variables_predictoras_numericas])
data.head()
```

```
data.describe()
```

```
#Creación de dummies a las variables categóricas

#Variables categóricas con 2 categorías -> Borramos una dummy
data = pd.get_dummies(data, columns=['Year'], drop_first=, dtype=int)

#Variables categóricas con más de 2 categorías -> No borramos
data = pd.get_dummies(data, columns=['Region'], drop_first=, dtype=int)
data.head()
```
