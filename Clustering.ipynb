```
#Importamos librerías básicas
import pandas as pd # manipulacion dataframes
import numpy as np  # matrices y vectores
import matplotlib.pyplot as plt #gráfica

```

# 1. Preparación de Datos











```
#Cargamos los datos
data = pd.read_excel("datos_limpios.xlsx", sheet_name=0)
data.head()
```

```
#Se revisan los tipos de datos
data.info()
```

```
#Corrección tipos de datos
data['Region']=data['Region'].astype('category')

# Completa las correcciones

data.info()
```

```
data['Year'].value_counts().plot(kind='bar')
```

```
# Solo se requieren datos activos, los más recientes son año 2022. Se debe tener un registro por país/persona/entidad
data = data[data['Year'] == 2022]
data['Year'].value_counts().plot(kind='bar')
```

```
#Eliminamos variables irrelevantes para el modelo de clustering

data = data.drop(columns='Unnamed: 0') # Id de fila irrelevante
data = data.drop(columns='Year') # Todos son del mismo año
data = data.drop(columns='Energías Renovables_t+1') # No se desea predecir
data = data.drop(columns='Nivel transicion energetica_t+1') # No se desea predecir

data.head()
```

```
#Histogramas de las variables numéricas
data.hist()
```

#Transformaciones: normalizar y crear dummies

```
#Transformaciones: normalizar y crear dummies

#Copia de los datos
data_original=data.copy()
```

```
#Normalizacion de variables numéricas
from sklearn.preprocessing import MinMaxScaler


variables_a_normalizar=['Petróleo_t','Gas Natural_t','Carbón_t','Energía Hidroeléctrica_t','Energías Renovables_t']

min_max_scaler = MinMaxScaler()

min_max_scaler.fit(data[variables_a_normalizar]) #Ajuste de parámetro
data[variables_a_normalizar]= min_max_scaler.transform(data[variables_a_normalizar])
data.head()
```

```
#Variables categóricas con más de 2 categorías -> No borramos
data = pd.get_dummies(data, columns=['Region'], drop_first= , dtype=int)
data.head()
```

# 2. Aprendizaje del Modelo
* Aplicar kmeans

```
#Creación de modelo de clustering con Kmeans
from sklearn.cluster import KMeans
k=4
model = KMeans(n_clusters=k, max_iter=300)
model.fit(data) #100% datos
```

# 3. Evaluación del Modelo


*   Inertia: medidas intra-cluster, valor pequeño esperado
*   Silueta: valor positivo esperado, idealmente mayor a 0.5



```
#Evaluación
from sklearn import metrics

#Inertia: se require valor pequeño
print('Inercia o cohesión:', model.inertia_)

#Silueta: se requiere que sea positivo, ideal 0.5-1.0
sil=metrics.silhouette_score(data, model.labels_)
print('Silueta:',sil)
```
